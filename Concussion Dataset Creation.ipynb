```python
import pandas as pd
```

```python
# read in the csv file containing the transaction data
df = pd.read_csv('injuries.csv')
```

```python
# filter data set on the condition that "concussion" appears in the notes column
concussions = df[df['Notes'].str.contains('concussion')]
concussions = concussions.reset_index(drop=True)
concussions
```

```python
# drop the 'Acquired' column and rename the 'Relinquised' column
concussions = concussions.drop(['Acquired'], axis=1)
concussions.rename(columns = {'Relinquised':'Player'}, inplace = True)

concussions
```

```python
# send the df to an excel file
concussions.to_excel('Concussions 2010-18.xlsx')
```
